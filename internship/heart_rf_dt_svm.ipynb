```python
import pandas as pd
```

```python
a=pd.read_csv('heart.csv')
a
```

```python
a.shape
```

```python
a.dropna(inplace=True)
```

```python
x=a.iloc[:,:13]
x
```

```python
y=a.iloc[:,13:]
y
```

```python
y=y["target"]
```

```python
from collections import Counter
Counter(y)
```

```python
from imblearn.over_sampling import RandomOverSampler
ros=RandomOverSampler()
x_ros,y_ros=ros.fit_resample(x,y)
```

```python
Counter(y_ros)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_ros,y_ros,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sclr=StandardScaler()
```

```python
x_train
```

```python
x_train=sclr.fit_transform(x_train)
```

```python
x_train
```

```python
x_test=sclr.fit_transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
predict=rf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
acc=accuracy_score(y_test,predict)*100
acc
```

```python
clr=classification_report(y_test,predict)
print(clr)
```

```python
con=confusion_matrix(y_test,predict)
print(con)
```

```python
import pickle
file=open("heart_sclr.pkl","wb")
pickle.dump(sclr,file)
file1=open("randfrt_heart.pkl","wb")
pickle.dump(rf,file1)
file.close()
file1.close()
```

```python
from sklearn.tree import DecisionTreeClassifier
clfdt=DecisionTreeClassifier()
clfdt.fit(x_train,y_train)
```

```python
ypred=clfdt.predict(x_test)
ypred
```

```python
acc=accuracy_score(y_test,ypred)*100
acc
```

```python
clr=classification_report(y_test,ypred)
print(clr)
```

```python
con=confusion_matrix(y_test,ypred)
print(con)
```

```python
import pickle 
file=open("heart_scalar.pkl",'wb')
pickle.dump(sclr,file)
file1=open("algorithmdt_heart.pkl",'wb')
pickle.dump(clfdt,file1)
file.close()
```
