```python
import pandas as pd
```

```python
a=pd.read_csv("framingham.csv")
a
```

```python
a=a.dropna()
a
```

```python
a=a.drop(["education"],1)
a
```

```python
x=a.iloc[:,:14]
y=a.iloc[:,14:]
y
```

```python
y=y['TenYearCHD']
```

```python
from collections import Counter
```

```python
Counter(y)
```

```python
from imblearn.over_sampling import RandomOverSampler
ros=RandomOverSampler()
x_ros,y_ros=ros.fit_resample(x,y)
```

```python
from imblearn.over_sampling import RandomOverSampler
ros=RandomOverSampler()
x_ros,y_ros=ros.fit_resample(x,y)
```

```python
Counter(y_ros)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_ros,y_ros,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sclr=StandardScaler()
```

```python
x_train=sclr.fit_transform(x_train)
```

```python
x_test=sclr.fit_transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
predict=rf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
acc=accuracy_score(y_test,predict)*100
acc
```

```python
clr=classification_report(y_test,predict)
print(clr)
```

```python
con=confusion_matrix(y_test,predict)
print(con)
```
