```python
import pandas as pd
```

```python
a=pd.read_csv("framingham.csv")
a
```

```python
a.shape
```

```python
a.dropna(inplace=True)
a
```

```python
a.shape
```

```python
a=a.drop(['education'],axis=1)
a
```

```python
z=a.iloc[:,0:14]
w=a.iloc[:,14:]
w
```

```python
w=w['TenYearCHD']
```

```python
from collections import Counter 
Counter(w)
```

```python
from imblearn.over_sampling import RandomOverSampler 
ros=RandomOverSampler()
x_ros,y_ros=ros.fit_resample(z,w)
```

```python
Counter(y_ros)
```

```python
#model building
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_ros,y_ros,test_size=0.2)
```

```python
x_train
```

```python
from sklearn.preprocessing import StandardScaler
sclr=StandardScaler()
```

```python
x_train=sclr.fit_transform(x_train)
```

```python
x_train
```

```python
x_test=sclr.fit_transform(x_test)
```

```python
x_test
```

```python
#svm algorithm
from sklearn.svm import SVC
clf=SVC(random_state=17)
#SVC(kernal='rbf',gamma='auto',class_weight='balanced',decision_function_shape='ovo',max_iter=-1,random_state=42)
clf.fit(x_train,y_train)
```

```python
ypred=clf.predict(x_test)
ypred
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
acc=accuracy_score(y_test,ypred)*100
acc
```

```python
clr=classification_report(y_test,ypred)
print(clr)
```

```python
con=confusion_matrix(y_test,ypred)
print(con)
```
